## K-NN Python

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore', category=UserWarning, append=True)
```

```python
url = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
df = pd.read_csv(url)
df
```

```python
x = df.iloc[:, :-1]
y = df.loc[:, "variety"]
y = df['variety'].values
```

```python
# splitting the data into training and test sets (80:20)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

### KNN Library python sklearn.neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
```

```python
x_new = [[5.8,	4.0,	1.2,	0.2]]
y_predict = knn.predict(x_new)

print(y_predict[0])
```

### KNN Tanpa library

```python
import numpy as np
def knn1(arr, x, wisata, k=1):
  knn = []
  for i in arr:
    dist = np.array([np.linalg.norm(x[j] - i) for j in range(len(x))])
    idx_k3n = np.where(dist <= dist[np.argsort(dist)])
    k3n = []
    for j in idx_k3n[0]:
      if len(k3n) < k:
        k3n.append(wisata[j])
    knn.append(k3n)
  return knn
```

```python
x = knn1(np.array(x_new),np.array(x_train), np.array(y_train))
x
```
