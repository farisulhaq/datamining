# Statistika Deskriptif

## Apa Itu Statistika Deskriptif?
<p align="justify">
Sebelum membahas pengertian Statistika Deskriptif lebih baik kita mencari tau terlebih dahulu apasih perbedaan statistik dengan statistika. 
<br>
Statistika mengacu pada salah satu cabang ilmu yang mempelajari tentang bagaimana mengumpulkan, mengolah, menganalisis, menyajikan, interpretasi data dan mengambil kesimpulan.
<br>
Sedangkan statistik adalah catatan angka dan hasil pengolahan data. Hasil pengolahan data bisa berupa rata-rata (mean), nilai tengah (median), modus dan masih banyak lagi. 
</p>

## Apa Kegunaan Statistika Deskriptif?
<p align="justify">
Salah satu kegunaan yang paling populer adalah untuk menganalisis dan mengolah data yang dikumpulkan dari sebuah penelitian.
<br>
Nah, statistika deskriptif ini berguna untuk memberikan informasi dan hubungan variabel dari data yang telah kamu kumpulkan dengan susah payah.
<br>
Selain itu, contoh statistika deskriptif  tetap bisa terpakai bahkan ketika kamu masuk ke dunia kerja sekalipun. Misalnya, kamu ingin memulai sebuah startup. Tentu kamu harus membuat perkiraan finansial usaha tersebut dari tahun ke tahun, termasuk pengeluaran dan pemasukannya. 
</p>

## Metode Statistika Deskriptif
<p align="justify">
Dengan menggunakan statistika deskriptif, kita bisa mendapatkan berbagai informasi dari data dan variabel sebagai berikut ini.
</p>

* <p align="justify">Pemusatan data pada data tunggal dan data berkelompok (mean, median, dan modus)</p>
* <p align="justify">Ukuran letak data tunggal dan data berkelompok (kuartil dan desil)</p>
* <p align="justify">Ukuran penyebaran data pada data tunggal (jangkauan, jangkauan antar kuartil, simpangan, simpangan rata-rata)</p>
* <p align="justify">Ukuran penyebaran data pada data berkelompok (jangkauan, jangkauan antar kuartil, simpangan kuartil, simpangan baku, dan ragam)</p>

### Mean, Median, dan Modus
<p align="justify">
Kalau kamu dari tadi bertanya-tanya tentang rumus statistika deskriptif, kamu perlu mengerti dulu nih tentang mean, median, dan modus.
</p>

* <p align="justify">Mean atau rata-rata adalah perbandingan jumlah seluruh data dan banyaknya data.</p>
* <p align="justify">Median adalah nilai tengah dari data yang ada. Gue pribadi punya cara sendiri untuk ngapalin apa itu median. Dari tulisannya kan awalnya ada ‘med’, seperti medium. Nah, medium itu biasanya tengah-tengah, di bawah large tapi di atas small. Jadi, gue selalu inget banget median itu artinya nilai tengah.</p>
* <p align="justify">Modus adalah nilai yang frekuensi kemunculannya paling banyak. Dengan kata lain, modus merupakan nilai yang paling sering muncul dibandingkan dengan nilai-nilai lainnya.</p>

### Kuartil, Desil, dan Persentil
* <p align="justify">kuartil adalah nilai yang membagi data menjadi 4 bagian secara sama rata.</p>
* <p align="justify">Desil adalah nilai yang membagi data menjadi 10 bagian secara sama rata.</p>
* <p align="justify">Persentil adalah nilai yang membagi data menjadi 100 bagian secara sama rata.</p>

### Jangkauan, Simpangan, Ragam, dan Varians
* <p align="justify">Jangkauan adalah nilai dari selisih nilai maksimum dan nilai minimum.</p>
* <p align="justify">Simpangan adalah Suatu nilai yang digunakan untuk mengukur seberapa jauh data yang lain menyimpang dari rata-rata.</p>
* <p align="justify">Ragam adalah nilai perbandingan antara simpangan baku (atau standar deviasi) dengan nilai rata-ratanya, dalam bentuk persen.</p>
* <p align="justify">Varians adalah nilai statistik yang sering kali dipakai dalam menentukan kedekatan sebaran data yang ada di dalam sampel dan seberapa dekat titik data individu dengan mean atau rata-rata nilai dari sampel itu sendiri.</p>

### Simpangan baku (Standar deviasi)
<p align="justify">
Standar deviasi atau biasa disebut simpangan baku merupakan nilai statistik yang menjelaskan tingkat homogenitas (atau keseragamanan) dari suatu data kelompok.
<br>
Nilai simpangan baku bisa berupa angka 0 atau selainnya:
</p>

* <p align="justify">Jika simpangan baku bernilai 0, itu menandakan setiap elemen dari suatu data kelompok bersifat indentik satu sama lain.</p>
* <p align="justify">Dan jika tidak sama dengan 0 (bisa negatif atau positif), itu menunjukkan bahwa titik data individu lebih kecil atau lebih besar dari rata-rata.</p>

## **Rumus**
### **Mean**
**Data Tunggal**

$$
\begin{align*}
\displaystyle\bar x &= \frac{x_{1} + x_{2} + x_{3} + ... + x_{n}}{n} \\
keterangan &: \\
\bar x &=\ Rata-rata \\
x_{1} + x_{2} + x_{3} + ... + x_{n} &=\ Nilai \  data \\
n &=\ Banyaknya \ data
\end{align*}
$$

**Data kelompok**
<p align="justify">
Mean untuk data kelompok Jika data yang tersedia merupakan data kelompok, artinya data itu dikelompokkan ke dalam interval-interval kelas yang sama panjang. Untuk mengetahui nilai mediannya dapat ditentukan dengan rumus berikut ini.
</p>

$$
\begin{align*}
\displaystyle
\bar x &= \frac{\sum f_{i}x_{i}}{n} \\
keterangan &: \\
\bar x &=\ Rata-rata \\
f_{i} &=\ Frekuensi \\
x_{i} &=\ Nilai \ tengah \\
n &=\ Total \ frekuensi \ atau \ data
\end{align*}
$$

### **Median**
**Data Tunggal :**

$$
\begin{align*}
\displaystyle
Jumlah \ data \ ganjil &: \\
M_{e} &= \frac{X_{(n -1)}}{2} \\
Jumlah \ data \ genap &: \\
M_{e} &= \frac{X_{(\frac{n}{2})}+X_{(\frac{n}{2}+1)}}{2}
\end{align*}
$$

**Data Kelompok**
<p align="justify">
Median untuk data kelompok Jika data yang tersedia merupakan data kelompok, artinya data itu dikelompokkan ke dalam interval-interval kelas yang sama panjang. Untuk mengetahui nilai mediannya dapat ditentukan dengan rumus berikut ini.
</p>

$$
\begin{align*}
\displaystyle
M_{e} &= L + c\left( \frac{\frac{1}{2}n - f_{k}}{\sum f_{i}} \right)
\end{align*}
$$

## **Modus**
<p align="justify">
Modus ialah nilai yang paling sering muncul atau nilai yang mempunyai frekuensi tertinggi. Jika suatu data hanya mempunyai satu modus disebut unimodal dan bila memiliki dua modus disebut bimodal, sedangkan jika memiliki modus lebih dari dua disebut multimodal. Modus dilambangkan dengan Mo.
</p>

**Data tunggal**
<p align="justify">
Modus dari data tunggal adalah data yang sering muncul atau data dengan frekuensi tertinggi. Perhatikan contoh soal berikut ini. Contoh: Tentukan modus dari data di bawah ini. 2, 1, 4, 1, 1, 5, 7, 8, 9, 5, 5, 10 Jawab: Data yang sering muncul adalah 1 dan 5. Jadi modusnya adalah 1 dan 5.
</p>

**Data Kelompok**
<p align="justify">
Modus data kelompok Modus data kelompok dirumuskan sebagai berikut:
</p>

$$
\begin{align*}
\displaystyle
M_{o} &= L + c\left( \frac{d_{1}}{d_{1} + d_{2}} \right)c \\
\end{align*}
$$


## **Desil**
<p align="justify">
Desil adalah membagi data yang telah diurutkan menjadi sepuluh bagian yang sama banyak. Desil data tunggal Urutkan data dari yang kecil ke yang besar, kemudian tentukan Desil dengan rumus sebagai berikut: 
</p>

$$
\begin{align*}
\displaystyle
D_{i} &= data \ ke -\frac{i\left( n + 1 \right)}{10}
\end{align*}
$$


## **Kuartil**
<p align="justify">
kuartil adalah membagi data yang telah diurutkan menjadi empat bagian yang sama banyak. Kuartil data tunggal Urutkan data dari yang kecil ke yang besar, kemudian tentukan kuartil dengan rumus sebagai berikut: 
</p>

$$
\begin{align*}
\displaystyle
Q_{i} &= \frac{i\left( n + 1 \right)}{4}
\end{align*}
$$


## **Jangkauan**
<p align="justify">
Apabila sekumpulan data sudah diurutkan dari yang terkecil sampai terbesar maka Range dari data itu adalah
<br>

**Range = data terbesar – data terkecil**

$$
\begin{align*}
\displaystyle
J &= X_{max} - X_{min}
\end{align*}
$$


Contohnya tentukan range dari data 15, 20, 25 , 35, 40, 43, 50!

Cara menjawabnya:
Data terbesarnya = 50
data terkecilnya = 15

**Range = 50 – 15 = 35**
</p>

## **Simpangan Baku**

$$
\begin{align*}
\displaystyle
Data \ Populasi &: \\
s &= \sqrt{\frac{\sum_{i=1}^{n}\left( x_{i} - \bar x \right)^{2}}{n}} \\
Data \ Sampel &: \\
s &= \sqrt{\frac{\sum_{i=1}^{n}\left( x_{i} - \bar x \right)^{2}}{n-1}}
\end{align*}
$$

## Implementasi


### Dataset 
Membaca dataset dari url menggunakan pandas

```python
import warnings
warnings.filterwarnings('ignore')
# import pandas
import pandas as pd
url="https://raw.githubusercontent.com/RamiKrispin/coronavirus-csv/master/south_korea/covid_south_korea_long.csv"
c = pd.read_csv(url)
```

```python
# mengecek type data pada dataset
c.info()
```

```python
# memilih data yang berupa numeric
df = c['total']
df
```

```python
# mengecek deskripsi dataset 
df.describe()
```

### Contoh Code
**Statistika Deskriptif**

```python
import numpy as np
import statistics as st
```

#### Mean

```python
x = st.mean(df)
print(f'Rata-rata = {x}')
```

#### Median

```python
Me = st.median(df)
print(f'Median = {Me}')
```

#### Modus

```python
Mo = st.mode(df)
print(f'Modus = {Mo}')
```

#### Kuartil

```python
# Kuartil ke-1
q1 = np.percentile(df, 25)
print('kuartil ke-1 = ', q1)
# Kuartil ke-2
q2 = np.percentile(df, 50)
print('kuartil ke-2 = ', q2)
# Kuartil ke-3
q3 = np.percentile(df, 75)
print('kuartil ke-3 = ', q3)
```

#### Desil

```python
D = np.percentile(df, np.arange(0, 100, 10))

for ind, data in enumerate(D):
  print(f'Desil ke-{ind+1} = {data}')
```

#### Jangkauan

```python
# rumus jamgkauan J = Xn - X1
Xn = max(df)
X1 = min(df)
J = Xn - X1
print(f'Jangkauan data = {J}')
```

#### Simpangan Baku

```python
s = st.pstdev(df)
print(f'Simpangan baku = {s}')
```

### Kemiringan/Skewness
<p align="justify">
Dalam teori probabilitas dan statistik, skewness adalah ukuran dari asimetri distribusi probabilitas dari variabel acak bernilai riil tentang nilai tengahnya. Nilai skewness bisa positif, nol, negatif, atau tidak terdefinisi.
</p>
