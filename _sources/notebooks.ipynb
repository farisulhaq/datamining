## read data with url in pandas

```python
import pandas as pd
url="https://raw.githubusercontent.com/RamiKrispin/coronavirus-csv/master/south_korea/covid_south_korea_long.csv"
c = pd.read_csv(url)
```

```python
c.info()
```

```python
data = c['total']
data
```

```python
data.describe()
```

```python
import numpy as np
np.percentile(data, 25)
```

```python
np.percentile(data, 50)
```

```python
np.percentile(data, 75)
```

```python
import statistics as st
```

## median

```python
st.median(data)
```

## modus

```python
st.mode(data)
```

## rata-rata

```python
st.mean(data)
```

## kemiringan/skewned

```python
data.skew()
```
