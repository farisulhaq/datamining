## Implementasi dengan K-NN

```python
import pandas as pd
import numpy as np
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/00451/dataR2.csv'

df = pd.read_csv(url)
df
```

### cek ukuran dataset 

```python
df.shape
```

### Cek missing value dataset

```python
df.isna().sum()
```

### data analysis

```python
df.info()
```

### memisah dataset

```python
X = df.iloc[:, :9]
y = df.iloc[:, 9]

X, y
```

### preprocessing data

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X=ss.fit_transform(X)
X=pd.DataFrame(X)
X
```

### menyemplit dataset dengan perbandingan data train 80% dan data tes 20%

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

### Model K-NN dengan sklearn

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

my_param_grid = {'n_neighbors':[2,3,5,7], 'weights': ['distance', 'uniform']}
GridSearchCV(estimator=KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
modelKNN = GridSearchCV(KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
```

```python
my_param_grid = {'n_neighbors':[2,3,5,7], 'weights': ['distance', 'uniform']}
GridSearchCV(estimator=KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
knn = GridSearchCV(KNeighborsClassifier(), param_grid=my_param_grid, refit=True, verbose=3, cv=3)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.best_params_
```

### karena nilai k terbaik dari uji coba yaitu k = 2, 3,5,7 maka kita bisa gunakan {'n_neighbors': 3, 'weights': 'distance'}

### data test predict

```python
test_pred = knn.predict(X_test)
```

### akurasi 

```python
print(classification_report(y_test, test_pred))
```

### inputan test

```python
y_pred = np.array([1.788816,	-0.080447,	1.792721,	0.987394,	1.125766,	3.333167,	0.576644,	-0.841036,	-1.290747,]).reshape(1,-1)
y_pred.shape
```

### melakukan predict terhadap inputan

```python
y_predict = knn.predict(y_pred)
```

```python
labels = ['Healthy controls', 'Patients']
```

```python
labels[y_predict[0]-1]
```

## Implementasi dengan Desision tree
